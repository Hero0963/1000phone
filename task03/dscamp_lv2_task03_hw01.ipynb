```python
import requests
from bs4 import BeautifulSoup as soup

url = "https://movies.yahoo.com.tw/movie_thisweek.html"
r = requests.get(url)
response = r.text


d = soup(r.text, "lxml")


movies_info = d.find_all("div", class_="release_info_text")
for movie_info in movies_info:
    # print(type(movie_info)
    # print(movie_info)
    chinese_name = movie_info.find("a", class_="gabtn").get_text().strip()
    eng_name = movie_info.find("div", class_="en").get_text().strip()
    release_time = movie_info.find("div", class_="release_movie_time").getText().replace('上映日期：','').strip()
    score = movie_info.find("div", class_="leveltext").get_text().replace('網友想看','').strip()
    display(chinese_name, eng_name, release_time, score)
  
```

```python
# ref = https://www.webscrapingpro.tw/yahoo-movie-web-scraping-using-python/

import requests
from bs4 import BeautifulSoup
import pandas as pd
 
url = 'https://movies.yahoo.com.tw/movie_thisweek.html'
response = requests.get(url=url)
 
soup = BeautifulSoup(response.text, 'lxml')

info_items = soup.find_all('div', 'release_info')

movies_info = []
for item in info_items:
    name = item.find('div', 'release_movie_name').a.text.strip()
    english_name = item.find('div', 'en').a.text.strip()
    release_time = item.find('div', 'release_movie_time').text.split('：')[-1].strip()
    expectation = item.find('div', 'leveltext').span.text.strip()

    movies_info.append([name, english_name, release_time, expectation])


pd_movies_info = pd.DataFrame(movies_info, columns=['中文名稱','英文名稱','上映日期','期待度'])
display(pd_movies_info)
```
